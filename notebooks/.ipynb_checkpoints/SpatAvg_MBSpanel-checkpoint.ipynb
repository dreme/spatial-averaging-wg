## Spatially averaged fields

#### Create some new S columns for vertical spatial averaging
c is the convolution vector applied to z columns of points for each (x,y) group

```python
def conv_vector(l, dl, m):
    '''Create a convolution vector
        l = length of convolution vector (m)
       dl = increment spacing between points in convolution vector (m)
        m = number of averaging points in convolution vector'''

    n = int(round(l/dl)) + 1  # number of points in convolution vector'''
    c = np.ones(n)            # convolution vector
    ix = np.arange(n)         # index of convolution vector
    dix = (n-1)/(m-1)         # index spacing between convolution averaging points

    assert np.isclose(l/dl, int(l/dl), rtol=1e-08), f'dl ({dl}) does not divide equally into l ({l})'
    assert type(m) == int, f'm ({m}) must be an integer'
    assert (n-1) % (m-1) == 0, f'm ({m}) does not distribute evenly across number of convolution points ({n})'

    c[(ix % dix) != 0] = 0
    c = c/m

    assert np.isclose(sum(c), 1, rtol=1e-08), f'sum of c ({sum(c)}) does not add up to 1'
    
    return c
```

```python
# 1.6m vertical average curves
for m in [17,5,3]:
    l, dl = 1.6, 0.1
    c = conv_vector(l, dl, m)
    print(f'\n{m:2d} points: sum{list(c.round(3))} = {sum(c)}')
    R.spatavg(f"SE-{l}m-{m}", "SE", c, title=f"FEKO SE levels ({m} points averaged over {l}m)")
    R.spatavg(f"Smax-{l}m-{m}",  "Smax",  c, title=f"FEKO Smax levels ({m} points averaged over {l}m)")  
```

## Thors $SAR_{10g}$ and $SAR_{wb}$ calcs

```python
def calcSAR10g(f,P,N,L,Φdeg,D0dBi,rlist):
    # constants
    A = 3 / 2
    B = 1
    π = np.pi
    λ = f / 300

    # convert Φ and D0 to correct units
    Φ = Φdeg * π / 180  # radians
    D0 = 10**(D0dBi/10) # linear

    # calculate r10g breakpoints
    r10g1 = λ / 4
    r10g2 = (Φ / 12) * D0 * L

    # Calculate SARlocal
    def cSAR10g(r):
        if r < r10g1:
            SAR10g = A * B * P / (N * r10g1)
        elif r <= r10g2:
            SAR10g = B * P / (N * Φ * r)
        else:
            SAR10g = B * P * r10g2 / (N * Φ * r**2)
        return SAR10g

    SAR10glist = list(map(cSAR10g,rlist))
    return SAR10glist   

def calcSARwb(f,P,Φdeg,D0dBi,rlist):
    # constants
    C = 1 / 75
    D = 1 / 200
    E = 8E8
    F = 2 / 5
    π = np.pi
    λ = f / 300

    # convert Φ and D0 to correct units
    Φ = Φdeg * π / 180  # radians
    D0 = 10**(D0dBi/10) # linear

    # calculate rwb breakpoints
    rwb1 = D / (C * Φ)
    if 800 <= f < 2000:
        rwb2 = E * D0 * Φ / (π * f * 1E6)
    elif 2000 <= f <= 2200:
        rwb2 = F * D0 * Φ / π
    else:
        raise NameError(f'f ({f}) must be between 800 and 2200 MHz')

    # Calculate SARwb
    def cSARwb(r):
        if r < rwb1:
            SARwb = C * P
        elif r <= rwb2:
            SARwb = D * P / (Φ * r)
        else:
            SARwb = D * P * rwb2 / (Φ * r**2)
        return SARwb

    SARwblist = list(map(cSARwb,rlist))
    return(SARwblist)
```

```python
# Test out the SAR functions
f = 900  # MHz
P = 80   # W
N = 9
L = 2.25 # m
Φdeg = 68 # degrees
D0dBi = 17  # dBi
xlist = np.array([0.5,1,2,3,4,5])
rlist = xlist - 0.04

SAR10glist = calcSAR10g(f,P,N,L,Φdeg,D0dBi,rlist)
SARwblist = calcSARwb(f,P,Φdeg,D0dBi,rlist)
pd.DataFrame(dict(x=xlist,r=rlist,SAR10g=SAR10glist,SARwb=SARwblist))
```

## Scratch

### Rolling maximum

```python
from numpy.lib.stride_tricks import sliding_window_view

def rolling_max(x,y,n):
    '''This function calculates the rolling maximum of the y array.
       It extrapolates values at the start and end of the y array 
       to accommodate the size of the rolling window.
       It assumes that the x values are ascending
    x = corresponding x values for the y array
    y = the array for which the rolling maximum is obtained
    n = size of the rolling window'''
    
    f = interpolate.interp1d(x, y, kind='quadratic',fill_value="extrapolate")
    dx = x[1] - x[0]
    m = int((n-1)/2)
    xlow = x[0:m] - dx*m
    xhigh = x[-m:] + dx*m
    ylow = f(xlow)
    yhigh = f(xhigh)
    y = np.insert(y,0,ylow)
    y = np.append(y,yhigh)
    ymax = np.max(sliding_window_view(y, window_shape = n), axis = 1)
    return ymax

x = np.arange(-5,6)
y = x**2
ymax = rolling_max(x,y,n)

df = pd.DataFrame(dict(x=x,y=y,ymax=ymax))
df.T
```

```python
from numpy.lib.stride_tricks import sliding_window_view

def rolling_max(S,z,n):
    '''This function calculates the rolling maximum of the S array.
       It extrapolates values at the start and end of the S array 
       to accommodate the size of the rolling window.
       It assumes that the z values are ascending
    S = the S array for which the rolling maximum is obtained
    z = corresponding z array for the S array
    n = size of the rolling window (must be an odd integer)'''
    
    # Create extrapolation function for S values
    f = interpolate.interp1d(z, S, kind='quadratic',fill_value="extrapolate")
    
    # Create additional z values for extrapolating S at start and end of array
    dz = z[1] - z[0]
    m = int((n-1)/2)
    z_start = z[0:m] - dz*m
    z_end   = z[-m:] + dz*m
    
    # Expand S array with start and end extrapolated values
    S_start = f(z_start)
    S_end   = f(z_end)
    S = np.insert(S, 0, S_start)
    S = np.append(S, S_end)
    
    # Calculate rolling maximum for S from extrapolated S array
    Smax = np.max(sliding_window_view(S, window_shape = n), axis = 1)
    
    return Smax

z = np.arange(-5,6)
S = z**2
n = 5
Smax = rolling_max(S,z,n)
display(Smax)

df = pd.DataFrame(dict(z=z,S=S,Smax=Smax))
df.T
```

```python
def draw_coordinate_frame_at_origin(fig):
    """
    Draw the origin and 3 vectors representing standard basis vectors to express
    a coordinate reference frame.
    Args:
       fig: Mayavi figure
    Returns:
       Updated Mayavi figure
    Based on
    --------
    https://github.com/hengck23/didi-udacity-2017/blob/master/baseline-04/kitti_data/draw.py
    https://github.com/charlesq34/frustum-pointnets/blob/master/mayavi/viz_util.py
    """
    # draw origin
    mlab.points3d(0, 0, 0, color=(1, 1, 1), mode="sphere", scale_factor=0.2)

    # Form standard basis vectors e_1, e_2, e_3
    axes = np.array([[2.0, 0.0, 0.0], [0.0, 2.0, 0.0], [0.0, 0.0, 2.0]], dtype=np.float64)
    # e_1 in red
    mlab.plot3d(
        [0, axes[0, 0]], [0, axes[0, 1]], [0, axes[0, 2]], color=(1, 0, 0), tube_radius=None, figure=fig
    )
    # e_2 in green
    mlab.plot3d(
        [0, axes[1, 0]], [0, axes[1, 1]], [0, axes[1, 2]], color=(0, 1, 0), tube_radius=None, figure=fig
    )
    # e_3 in blue
    mlab.plot3d(
        [0, axes[2, 0]], [0, axes[2, 1]], [0, axes[2, 2]], color=(0, 0, 1), tube_radius=None, figure=fig
    )

    return fig 


fig = mlab.figure(1, size=(1200, 900), bgcolor=(0.5, 0.5, 0.5))
mlab.clf()

draw_coordinate_frame_at_origin(fig)
mlab.show()
```

#### find maximum method

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Enter input parameters for distributions of z and spatially averaged S 
h = 2          # height of the spatial averaging length
dz = 0.1       # z point increment
zlow = 0       # lower bound of z points
zhigh = 20     # upper bound of z points
zsource = 10   # z level of RF source

# Generate distributions of z and spatially averaged S
nz = int((zhigh-zlow)/dz+1)     # number of z points
z = np.linspace(zlow,zhigh,nz)  # z distribution
k1 = 1
k2 = 1
Ssa = k1/np.cosh(k2*(z-zsource))   # artificial distribution of spatial average S

# Count number of z points in averaging length and half averaging length
nh = int(h/dz + 1)       # number of z points in averaging length
ni = int((nh - 1) / 2)   # number of z points to shift array values

# create point spatial distribution for point spatial S
imax = np.argmax(Ssa)  # array index for maximum value of Ssa
zlow = z[ni:imax+1] 
zhigh = z[imax:-ni]
zps = np.append(zlow,zhigh)

Sps_low = Ssa[:imax-ni+1]
Sps_high = Ssa[imax+ni:]
Sps = np.append(Sps_low,Sps_high)

# plot the distributions
plt.figure(figsize=(3,6))
plt.plot(Ssa, z, alpha=0.7, label='spatial average')
plt.plot(Sps, zps, alpha=0.7, label='point spatial')
plt.xlabel('S (W/m²)')
plt.ylabel('z (m)')
plt.legend()
pass
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib qt

# Enter input parameters for distributions of z and spatially averaged S 
h = 2          # height of the spatial averaging length
dz = 0.1       # z point increment
zlow = 0       # lower bound of z points
zhigh = 20     # upper bound of z points
zsource = 10   # z level of RF source

# Generate distributions of z and spatially averaged S
nz = int((zhigh-zlow)/dz+1)     # number of z points
z = np.linspace(zlow,zhigh,nz)  # z distribution
k1 = 1
k2 = 1
Ssa1 = k1/np.cosh(k2*(z-zsource+3))   # artificial distribution of spatial average S
Ssa2 = k1/np.cosh(k2*(z-zsource-3))   # artificial distribution of spatial average S
Ssa = Ssa1 + Ssa2

# Count number of z points in averaging length and half averaging length
nh = int(h/dz + 1)       # number of z points in averaging length
ni = int((nh - 1) / 2)   # number of z points from centre of averaging length to either end

# create point spatial distribution for z
Sps = np.repeat(np.nan,len(z)) # initialise Sps array with Nan's
for i in range(ni,nz-ni):
    if Ssa[i+1] > Ssa[i-1]:
        Sps[i] = Ssa[i-ni]
    else:
        Sps[i] = Ssa[i+ni]
#     print(f'{i:2g}  {z[i]:2g}  {Ssa[i]:5.4f}  {Sps[i]:5.4f}')

# plot the distributions
plt.figure(figsize=(4,8))
plt.plot(Ssa, z, alpha=0.7, label='spatial average')
plt.plot(Sps, z, alpha=0.7, label='point spatial')
plt.xlabel('S (W/m²)')
plt.ylabel('z (m)')
plt.grid()
plt.legend()
plt.yticks(range(0,21))
plt.ylim((0,20))
plt.tight_layout()
plt.savefig('C:\\Users\\emeau\\OneDrive\\Desktop\\saps.png',dpi=300)
pass
```

#### Plot ps, sa, pssa

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib qt

# Enter input parameters for spatial averaging scheme 
h    = 2    # height of the spatial averaging window
nsap = 3    # number of spatial averaging points [3 or 5]
assert nsap in [3,5], f'nsap ({nsap}) must be 3 or 5'

# Enter input parameters for vertical distance points (z) 
zlow = 0    # lower bound of z points
zhigh = 20  # upper bound of z points
dz = 0.1    # z point increment

# Enter input parameters for fabricated point spatial distribution of limit normalised values (Lps) 
k1 = 1     # peak level of Lps distribution
k2 = 1     # vertical beam width parameter of Lps distribution
zL1 = 7    # z level of source 1
zL2 = 13   # z level of source 2

# Generate z distribution
nz = int((zhigh-zlow) / dz+1)     # number of z points
z = np.linspace(zlow, zhigh, nz) 

# Generate Lps distribution
Lps1 = k1 / np.cosh(k2*(z-zL1))   # artificial distribution of Lps for source 1
Lps2 = k1 / np.cosh(k2*(z-zL2)) * 0  # artificial distribution of Lps for source 2
Lps = Lps1 + Lps2                 # combined Lps distribution of source 1 and 2

# Generate Lps_local distribution
Lps_local = Lps / 4

# Calculate spatial average of Lps (Lsa) 
nh = int(h/dz + 1)       # number of z points within averaging window
ni = int((nh - 1) / 2)   # number of z points from centre of averaging window to either end

errmsg = f'Change h ({h}), nsap ({nsap}), or dz ({dz}) so that spatial averaging points align with z distribution points'
assert 2*ni*dz == h, errmsg  

if nsap == 3:
    hpoints = [-ni, 0, ni]
elif nsap == 5:
    ni2 = int(ni/2)
    assert ni == ni2 * 2, errmsg
    hpoints = [-ni, -ni2, 0, ni2, ni]
    
Lsa = np.repeat(np.nan, nz)  # initialise Lsa array with NaN's (Not a Number)
for iz in range(ni, nz-ni):
    h_values = [Lps[iz+ih] for ih in hpoints]  # Lps values to be averaged
    Lsa[iz] = np.mean(h_values) 
    
# create point spatial distribution for z
Lpssa = np.repeat(np.nan,len(z)) # initialise Lpssa array with Nan's
for iz in range(2*ni, nz-2*ni):
    if Lsa[iz+1] > Lsa[iz-1]:
        Lpssa[iz] = Lsa[iz-ni]
    else:
        Lpssa[iz] = Lsa[iz+ni]
        
# plot the Lps, Lsa & Lpssa distributions
plt.figure(figsize=(4,8))
plt.plot(Lps, z, alpha=0.7, label='point spatial')
plt.plot(Lsa, z, alpha=0.7, label='spatial average')
plt.plot(Lpssa, z, alpha=0.7, label='point spatial spatial average')
plt.plot(Lps_local, z, alpha=0.7, label='point spatial local')
plt.xlabel('Limit normalised field value')
plt.ylabel('z (m)')
plt.grid()
plt.legend()
plt.yticks(range(0,21))
plt.ylim((0,20))
plt.tight_layout()
title = f'{nsap} points over {h}m'
plt.title(title)
plt.savefig('C:\\Users\\emeau\\OneDrive\\Desktop\\saps.png',dpi=300)
plt.show()
pass
```
