### Plot fitted values

#### Relative permittivity

```python
# Calculate fitted values
fMHz = np.logspace(np.log10(10),np.log10(7000),100)
er_calc = fn_er(fMHz)

# Plot values
fig, ax = plt.subplots(figsize=(6,1.5))
ax.plot(fMHz_er,er,'o',markersize=3)
ax.plot(fMHz,er_calc,'-')
plt.grid(which='both')
plt.xscale('log')
plt.yscale('log')
plt.title('Relative permittivity for medium dry ground')
plt.xlabel('Freq (MHz)')
plt.ylabel('$\epsilon_r$')
```

#### Conductivity

```python
# Calculate fitted values
fMHz = np.logspace(np.log10(10),np.log10(7000),100)
σ_calc = fn_σ(fMHz)

# Plot values
fig, ax = plt.subplots(figsize=(6,5))
ax.plot(fMHz_σ,σ,'o',markersize=3)
ax.plot(fMHz,σ_calc,'-')
plt.grid(which='both')
plt.xscale('log')
plt.yscale('log')
plt.title('Conductivity for medium dry ground')
plt.xlabel('Freq (MHz)')
plt.ylabel('$\sigma$ (S/m)')
```

### Calculate dielectric values for sptial averaging project

```python
FREQS = [30, 60, 100, 300, 600, 1000, 3000]
er_fit = fn_er(FREQS).round(1)
σ_fit = fn_σ(FREQS).round(4)
df = pd.DataFrame({'fMHz':FREQS, 'er':er_fit, 'σ':σ_fit})
df.to_clipboard()
df
```
