# Spatial averaging techniques

$\Large S_{avg}(z_0) = \frac{\int_{z_0-L/2}^{z_0+L/2} S\ dz}{L}$

## Setup

```python
# import other modules
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
from scipy import interpolate
from icecream import ic

# Set plotting environment
sn.set()

# Set plot file directory
%cd -0
%cd "..\plots\"
```

## Define function for S(z)

```python
# Define function for S
@np.vectorize
def fn(x):
    y = 1 + 1/(x+1) + x**3
    return y
```

## Plot S(z)

```python
# Plots 
zlow, zhigh = 0,1.6
z1 = np.arange(zlow,zhigh+0.01,0.01)
y1 = fn(z1)
z2 = [0,0.4,0.8,1.2,1.6]
y2 = fn(z2)

fig, ax = plt.subplots(figsize=(6,4,))
ax.plot(z1,y1)
ax.plot(z2,y2,'ro')
for i, (z,y) in enumerate(zip(z2,y2)):
    j = i + 1
    text = f'$S_{j}$'
    ax.text(z,y+0.3,text,horizontalalignment='center',color='b')
for z in [0.2,0.6,1,1.4]:
    ax.text(z,1.1,'$\Delta z$',horizontalalignment='center')
    ax.arrow(z, 1, 0.2, 0,head_length=0.05,head_width=0.1,
             length_includes_head=True, color='black',lw=0.3)
    ax.arrow(z, 1, -0.2, 0,head_length=0.05,head_width=0.1,
             length_includes_head=True, color='black',lw=0.3)
ax.set_xticks(np.arange(0,1.8,0.2))
ax.set_xlim(zlow-0.2,zhigh+0.2)
ax.set_ylim(bottom=0, top=6)
ax.set_xlabel('z (m)',fontsize=14)
ax.set_ylabel('S (W/m²)',fontsize=14)
ax.set_title(' S function plot with 5 sample points',color='green', fontsize=18)
fig.tight_layout()
fig.savefig('spat avg function plot.png', dpi=100)                       
```

## Simple average approach 

$\large
\begin{split}
S_{avg} & = \frac{[S_1+S_2+S_3+S_4+S_5]}{5} \\
& =\frac{\Delta z [S_1+S_2+S_3+S_4+S_5]}{L+\Delta z}
\end{split}$

```python
zlow, zhigh = 0,1.6
z1 = np.arange(zlow,zhigh+0.1,0.1)
y1 = fn(z1)
z2 = [0,0.4,0.8,1.2,1.6]
y2 = fn(z2)

fig, ax = plt.subplots(figsize=(6,4,))
ax.plot(z1,y1)
ax.plot(z2,y2,'ro')
ax.bar(z2,y2,width=0.4, alpha=0.5, color='r')

for i, (z,y) in enumerate(zip(z2,y2)):
    j = i + 1
    text = f'$S_{j}$'
    ax.text(z,y+0.3,text,horizontalalignment='center',color='b')

ax.set_xticks(np.arange(0,1.8,0.2))
ax.set_xlim(zlow-0.2,zhigh+0.2)
ax.set_ylim(bottom=0, top=6)
ax.set_xlabel('z (m)',fontsize=14)
ax.set_ylabel('S (W/m²)',fontsize=14)
ax.set_title('Simple average of points',color='green', fontsize=18)
fig.tight_layout()
fig.savefig('spat avg simple average.png', dpi=100)                                              
```

## Riemann sum (mid point rule)

$\large
\begin{split}
S_{avg} & = \frac{\Delta z [S_{12}+S_{23}+S_{34}+S_{45}]}{L} \\
& = \frac{[S_{12}+S_{23}+S_{34}+S_{45}]}{4}\quad \text{since } \Delta z = L/4
\end{split}$

```python
zlow, zhigh = 0,1.6
z1 = np.arange(zlow,zhigh+0.1,0.1)
y1 = fn(z1)
z2 = [0.2,0.6,1.0,1.4]
y2 = fn(z2)

fig, ax = plt.subplots(figsize=(6,4,))
ax.plot(z1,y1)
ax.plot(z2,y2,'ro')
ax.bar(z2,y2,width=0.4, alpha=0.5, color='r')

for i, (z,y) in enumerate(zip(z2,y2)):
    j = f'{i+1}{i+2}'
    text = '$S_{' + f'{j}' + '}$'
    ax.text(z,y+0.3,text,horizontalalignment='center',color='b')

ax.set_xticks(np.arange(0,1.8,0.2))
ax.set_xlim(zlow-0.2,zhigh+0.2)
ax.set_ylim(bottom=0, top=6)
ax.set_xlabel('z (m)',fontsize=14)
ax.set_ylabel('S (W/m²)',fontsize=14)
ax.set_title('Riemann Sum (mid point rule)',color='green', fontsize=18)
fig.tight_layout()
fig.savefig('spat avg Riemann sum.png', dpi=100)                                              
```

## Trapezoidal rule

$\large
\begin{split}
S_{avg} &= \frac{\frac{\Delta z}{2}\enspace[S_1+2S_2+2S_3+2S_4+S_5]}{L} \\
&= \frac{[S_1 + 2S_2 + 2S_3 + 2S_4 + S_5]}{8}\quad \text{since } \Delta z = L/4
\end{split}$

```python
zlow, zhigh = 0,1.6
z1 = np.arange(zlow,zhigh+0.1,0.1)
y1 = fn(z1)
z2 = [0,0.4,0.8,1.2,1.6]
y2 = fn(z2)

fig, ax = plt.subplots(figsize=(6,4,))
ax.plot(z1,y1)
ax.plot(z2,y2,'ro')
ax.fill_between(z2,y2,color= "r",alpha= 0.5)

for i, (z,y) in enumerate(zip(z2,y2)):
    j = i + 1
    text = f'$S_{j}$'
    ax.text(z,y+0.3,text,horizontalalignment='center',color='b')

ax.set_xticks(np.arange(0,1.8,0.2))
ax.set_xlim(zlow-0.2,zhigh+0.2)
ax.set_ylim(bottom=0, top=6)
ax.set_xlabel('z (m)',fontsize=14)
ax.set_ylabel('S (W/m²)',fontsize=14)
ax.set_title('Trapezoidal Rule',color='green', fontsize=18)
fig.tight_layout()
fig.savefig('spat avg trapezoidal rule.png', dpi=100)                                              
```

## Simpson 1/3 rule

$\large
\begin{split}
S_{avg} &= \frac{\frac{\Delta z}{3}\enspace[S_1 + 4S_2 + 2S_3 + 4S_4 + S_5]}{L} \\
&= \frac{[S_1 + 4S_2 + 2S_3 + 4S_4 + S_5]}{12}\quad \text{since } \Delta z = L/4
\end{split}$

```python
zlow, zhigh = 0,1.6
z1 = np.arange(zlow,zhigh+0.1,0.1)
y1 = fn(z1)
z2 = [0,0.4,0.8,1.2,1.6]
y2 = fn(z2)
fi = interpolate.interp1d(z2, y2, kind='quadratic')
y3 = fi(z1)

fig, ax = plt.subplots(figsize=(6,4,))
ax.plot(z1,y1)
ax.plot(z2,y2,'ro')
ax.fill_between(z1,y3,color= "r",alpha= 0.5)

for i, (z,y) in enumerate(zip(z2,y2)):
    j = i + 1
    text = f'$S_{j}$'
    ax.text(z,y+0.3,text,horizontalalignment='center',color='b')

ax.set_xticks(np.arange(0,1.8,0.2))
ax.set_xlim(zlow-0.2,zhigh+0.2)
ax.set_ylim(bottom=0, top=6)
ax.set_xlabel('z (m)',fontsize=14)
ax.set_ylabel('S (W/m²)',fontsize=14)
ax.set_title("Simpson's 1/3 Rule",color='green', fontsize=18)
fig.tight_layout()
fig.savefig('spat avg Simpson one third rule.png', dpi=100)                                              
```

## Simpson 3/8 rule

$\large
\begin{split}
S_{avg} &= \frac{\frac{3\Delta z}{8}\enspace[S_1 + 3S_2 + 3S_3 + S_4]}{L} \\
&= \frac{[S_1 + 3S_2 + 3S_3 + S_4]}{8}\quad \text{since } \Delta z = L/3
\end{split}$

```python
zlow, zhigh = 0,1.6
z1 = np.arange(zlow,zhigh+0.1,0.1)
y1 = fn(z1)
z2 = [0,1.6/3,3.2/3,1.6]
y2 = fn(z2)
fi = interpolate.interp1d(z2, y2, kind='cubic')
y3 = fi(z1)

fig, ax = plt.subplots(figsize=(6,4,))
ax.plot(z1,y1)
ax.plot(z2,y2,'ro')
ax.fill_between(z1,y3,color= "r",alpha= 0.5)

for i, (z,y) in enumerate(zip(z2,y2)):
    j = i + 1
    text = f'$S_{j}$'
    ax.text(z,y+0.3,text,horizontalalignment='center',color='b')

ax.set_xticks(np.arange(0,1.8,0.2))
ax.set_xlim(zlow-0.2,zhigh+0.2)
ax.set_ylim(bottom=0, top=6)
ax.set_xlabel('z (m)',fontsize=14)
ax.set_ylabel('S (W/m²)',fontsize=14)
ax.set_title("Simpson's 3/8 Rule",color='green', fontsize=18)
fig.tight_layout()
fig.savefig('spat avg Simpson three eigths rule.png', dpi=100)                                              
```

## Gaussian Legendre quadrature rule

$\begin{split}
S_{avg} &= \frac{h\ [\text{0.236927}S_1 + \text{0.478629}S_2 + \text{0.568889}S_3 + \text{0.478629}S_4 + \text{0.236927}S_5]}{L} \\
&= \frac{[\text{0.236927}S_1 + \text{0.478629}S_2 + \text{0.568889}S_3 + \text{0.478629}S_4 + \text{0.236927}S_5]}{2}\quad \text{since } h = L/2
\end{split}$

```python
# Gauss points and weights

zlow, zhigh = 0,1.6

# Gauss points
zr = [-0.906180,-0.538469,0,0.538469,0.906180]
w = [0.236927,0.478629,0.568889,0.478629,0.236927]
z2 = (zlow+zhigh)/2 + (zhigh-zlow)*(np.array(zr)/2)
y2 = fn(z2)

# function curve
z1 = np.arange(zlow,zhigh+0.1,0.1)
y1 = fn(z1)

z4 = [0,0.4,0.8,1.2,1.6]
y4 = fn(z4)
fi = interpolate.interp1d(z4, y4, kind='cubic')
y3 = fi(z1)

fig, ax = plt.subplots(figsize=(6,4,))
ax.plot(z1,y1)
ax.plot(z2,y2,'ro')
ax.fill_between(z1,y3,color= "r",alpha= 0.5)

for i, (z,y) in enumerate(zip(z2,y2)):
    j = i + 1
    text = f'$S_{j}$'
    ax.text(z,y+0.3,text,horizontalalignment='center',color='b')

ax.set_xticks(np.arange(0,1.8,0.2))
ax.set_xlim(zlow-0.2,zhigh+0.2)
ax.set_ylim(bottom=0, top=6)
ax.set_xlabel('z (m)',fontsize=14)
ax.set_ylabel('S (W/m²)',fontsize=14)
ax.set_title("Gaussian Legendre Quadrature Rule",color='green', fontsize=18)
fig.tight_layout()
fig.savefig('spat avg Gauss quadrature rule.png', dpi=100)                                              
```
